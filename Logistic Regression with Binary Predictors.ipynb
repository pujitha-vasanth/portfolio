### Load the libraries

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

### Load the train dataset

```python
raw_data = pd.read_csv("Binary_predictors.csv")
raw_data
```

Let's change the categorical values to 0s and 1s. 

```python
# Map the train dataset
data = raw_data
data['Admitted'] = data['Admitted'].map({'Yes': 1, 'No': 0})
data['Gender'] = data['Gender'].map({'Female': 1, 'Male': 0})
data
```

### Assign the dependent and independent variables

```python
y = data['Admitted']
x1 = data[['SAT','Gender']]
```

```python
x = sm.add_constant(x1)
reg_log = sm.Logit(y,x)
results_log = reg_log.fit()

# Get the regression summary
results_log.summary()
```

The LLR p value ~0.000, and so the model is significant. The variable gender and SAT are also significant (small p-values).Using the logit equation, after calculating the odds of a female getting addmitted to the odds of a male getting admitted, a simplified equation was created, for which we will not calculate the exponential value of 1.9449. 

```python
np.exp(1.9449)
```

### Calculate the accuracy of the model

```python
# This is a method to change the formatting of np arrays
np.set_printoptions(formatter={'float': lambda x: "{0:0.2f}".format(x)})

results_log.predict()
```

```python
# An array containing the TRUE (actual) values
np.array(data['Admitted'])
```

```python
# A prediction table (confusion matrix) showing the 
results_log.pred_table()
```

```python
# Some neat formatting to read the table (better when seeing it for the first time)
cm_df = pd.DataFrame(results_log.pred_table())
cm_df.columns = ['Predicted 0','Predicted 1']
cm_df = cm_df.rename(index={0: 'Actual 0',1:'Actual 1'})
cm_df
```

```python
# Create an array (so it is easier to calculate the accuracy)
cm = np.array(cm_df)
# Calculate the accuracy of the model
accuracy_train = (cm[0,0]+cm[1,1])/cm.sum()
accuracy_train
```

The accuracy of the model with the train data is 94.6%. The dataset was initially divided into two, with a 90:10 ratio, as train and test datasets. This was done to resolve the problem of overfitting. Now, we will test the model with the test dataset. 

### Testing the model and assessing its accuracy

```python
# Load the test dataset
test = pd.read_csv('Test_dataset.csv')
test
```

```python
# Map the test data, as done with the train data
test['Admitted'] = test['Admitted'].map({'Yes': 1, 'No': 0})
test['Gender'] = test['Gender'].map({'Female': 1, 'Male': 0})
test
```

```python
# Let's see how the inputs from our x variable look
x
```

We see that there are 3 columns- const, SAT, and gender, in that order. Order is very important because the co-efficients of regression will expect it. If we fail to deliver in the correct order, the prediction will be wrong. 

```python
# Create a new variable called test_actual to put in the admission information (the actual outcomes that were observed)
test_actual = test['Admitted']
```

```python
# Create another variable called test_data which will contain everything from the test dataset except the admission information
test_data = test.drop(['Admitted'],axis=1)

# Use the add_constant from statsmodels to add a constant to the variable
test_data = sm.add_constant(test_data)
test_data
```

Now the data looks exactly like the input data. We got a bit lucky here as the variables remain in the same order. We need not change anything else. 

```python
def confusion_matrix(data,actual_values,model):
        
        # Confusion matrix 
        
        # There are three parameters
        # ----------
        # data: data frame or array
            # data is a data frame formatted in the same way as your input data (without the actual values)
            # e.g. const, var1, var2, etc. Order is very important!
        # actual_values: data frame or array
            # These are the actual values from the test_data
            # In the case of a logistic regression, it should be a single column with 0s and 1s
            
        # model: a LogitResults object
            # this is the variable where you have the fitted model 
            # e.g. results_log in this course
        # ----------
        
        #Predict the values using the Logit model
        pred_values = model.predict(data)
        # Specify the bins 
        bins=np.array([0,0.5,1])
        # Create a histogram, where if values are between 0 and 0.5 tell will be considered 0
        # if they are between 0.5 and 1, they will be considered 1
        cm = np.histogram2d(actual_values, pred_values, bins=bins)[0]
        # Calculate the accuracy
        accuracy = (cm[0,0]+cm[1,1])/cm.sum()
        # Return the confusion matrix and the accuracy
        return cm, accuracy
```

```python
# Create a confusion matrix with the test data
cm = confusion_matrix(test_data,test_actual,results_log)
cm
```

We see that the test data accuracy is 89.4%. Our train accuracy will always be higher than the test accuracy, as seen, because of overfitting. The regression fitted the training data as well as possible, but that does not mean that the prediction is true for all the values of the population. That's why we test data the model has never seen before. 

```python
# Format for easier understanding (not needed later on)
cm_df = pd.DataFrame(cm[0])
cm_df.columns = ['Predicted 0','Predicted 1']
cm_df = cm_df.rename(index={0: 'Actual 0',1:'Actual 1'})
cm_df
```
